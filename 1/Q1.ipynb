```
from scipy.io import loadmat
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.manifold import TSNE
import plotly.express as px

def Q1_a():
    data1 = loadmat(r'/content/drive/My Drive/Projects/ML_Assignment/dataset_1.mat')

    data1_label = data1['labels'].reshape(50000,1)
    data1_image = data1['samples'].reshape(50000,784)

    #print(list(np.unique(data1_label)))

    fig, ax = plt.subplots(nrows=10, ncols=10,figsize=(20,10))
    p = list(np.unique(data1_label))
    for i in p:
        k = 1
        j=0
        while(k<11 and j<data1_image.shape[0]):
            if(data1_label[j]==i):
                ax[p.index(i),k-1].imshow(data1_image[j].reshape(28,28), cmap="gray") 
                #plt.show()
                k = k+1
            j=j+1
            

Q1_a()

def Q1_b():
    data2 = loadmat(r'/content/drive/My Drive/Projects/ML_Assignment/dataset_2.mat')
    
    #print(data2['samples'].shape)
    #print(data2['labels'].shape)
    
    data2_label = data2['labels'].reshape(20000,1)
    data2_image = data2['samples'].reshape(20000,2)

    #print(list(np.unique(data2_label)))
    #print(data2_image[10].shape)

    df2_i = pd.DataFrame(data2_image,columns=['1','2'])
    #print(df2_i.head(4))
    df2_l = pd.DataFrame(data2_label,columns=['Label'])
    
    df2 = pd.concat([df2_i,df2_l],axis=1)
    
    plt.figure(figsize=(10,5))
    sn.set(style = 'whitegrid')
    sn.scatterplot(data= df2,x='1',y='2',hue='Label')
    plt.show()
            
Q1_b()            
  
def Q1_c():
    data1 = loadmat(r'/content/drive/My Drive/Projects/ML_Assignment/dataset_1.mat')

    data1_label = data1['labels'].reshape(50000,1)
    data1_image = data1['samples'].reshape(50000,784)
  
    tsne = TSNE(n_components=2)
    df1_i_2 = tsne.fit_transform(data1_image)
    
    df1_tsne_1 = pd.concat([pd.DataFrame(df1_i_2,columns=['1','2']) ,pd.DataFrame(data1_label,columns=['Label'])],axis=1)
    
    plt.figure(figsize=(10,5))
    sn.set(style = 'whitegrid')
    sn.scatterplot(data= df1_tsne_1,x='1',y='2',hue='Label',palette="deep")
    plt.legend()
    plt.show()

Q1_c()

def Q1_d():
    
    data1 = loadmat(r'/content/drive/My Drive/Projects/ML_Assignment/dataset_1.mat')

    data1_label = data1['labels'].reshape(50000,1)
    data1_image = data1['samples'].reshape(50000,784)

    print(data1_image.shape)
    tsne = TSNE(n_components=3)
    df1_i_2 = tsne.fit_transform(data1_image)
    print(df1_i_2.shape)
    df1_tsne_1 = pd.concat([pd.DataFrame(df1_i_2,columns=['1','2','3'])
    ,pd.DataFrame(data1_label,columns=['Label'])],axis=1)
     
    fig = px.scatter_3d(df1_tsne_1, x='1', y='2', z='3', color='Label')
    fig.show()
    plt.show()
    
Q1_d()
    
              
            
            

        
        
        
        
    
```
